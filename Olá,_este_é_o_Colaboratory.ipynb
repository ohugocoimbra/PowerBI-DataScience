```
import imageio
import imgaug as ia
%matplotlib inline

imagesInput = [
               imageio.imread("/content/gdrive/MyDrive/Fotos/augmentation/sanMateoBPack/input_modified/sanMateoBPack9_n.jpg")
]

print ("Original")
for i in range(len(imagesInput)):
  ia.imshow(imagesInput[i])
```

```
import imgaug.augmenters as iaa
import numpy as np

seq = iaa.Sequential([
                      #iaa.GaussianBlur(sigma=(0, 1.0)),
                      #iaa.AdditiveGaussianNoise(scale=(50, 60)),
                      #iaa.Crop(percent=(0, 0.2), keep_size=True),
                      #iaa.ElasticTransformation(alpha=(0, 10.0), sigma=0.25),
                      iaa.Dropout(p=(0, 0.15)),
                      #iaa.Fliplr(1),

                      #iaa.KeepSizeByResize(
                          #iaa.Crop((100,300), keep_size=True)
                      #),

                      iaa.Affine(
                          #scale={"x": (0.8, 1.2), "y":(0.8, 1.2)},
                          shear = (-8,8),
                          #translate_percent={"x": (-0.2, 0.2), "y":(-0.2, 0.2)}
                          #rotate=(-1,1),
                      )

], random_order=True)

images_aug = seq(images=imagesInput)

print("Augmented:")
for i in range(len(imagesInput)):
  ia.imshow(np.hstack([imagesInput[i], images_aug[i]])) #[imagesInput[i],]
```
